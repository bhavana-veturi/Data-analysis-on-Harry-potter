```python
import nltk


character_counter = dict()

books = list([1, 2, 3, 4, 5, 6, 7])
entire_text = ''
for i in books:
    print('Book {}'.format(i))
    book_name = "book_%s.txt" % (str(i))
    with open(('book_%s.txt' % (str(i))), 'r',encoding='utf-8') as file:
        contents = file.read()
        entire_text +=contents
book = (nltk.word_tokenize(entire_text))

```

```python
mytext = nltk.Text(book)
words = ["Harry", "Ron", "Hermione", "Hagrid","Malfoy", "Snape", "Dumbledore", "Voldemort"]
nltk.draw.dispersion.dispersion_plot(mytext, words, ignore_case=False,title='Word Dispersion in the Harry Potter Series')
```
